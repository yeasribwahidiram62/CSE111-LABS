```python
def replace_domain(email, new_domain, old_domain = "n/a"):
    if old_domain == "n/a":
        old_domain = email.split("@")[1]
    if old_domain == email.split("@")[1]:
        x = email.split("@")[0]
        new_mail = f"{x}@{new_domain}"
        print(f"Changed: {new_mail}")
    else:
        print(f"Unchanged: {email}")
email= "alice@kaaj.xyz"
new_domain= "sheba.xyz"
old_domain = "kaaj.com"
replace_domain(email, new_domain, old_domain = "n/a")
```

```python
def is_palindrome(input):
    y = ""
    for i in input:
        x = input.split()
        if i != " ":
            y += i
    for i in range(len(y)//2):
        if y[i] != y[-i-1]:
            return False
    return True
input = "nurses run"
if is_palindrome(input):
    print("Palindrome")
else:
    print("Not a Palindrome")
```
