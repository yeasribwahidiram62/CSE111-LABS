```python

unique=[]
freq=[]
while True:
    user = input()
    if user =="STOP":
        break
    num = int(user)

    if num in unique:
        index= unique.index(num)
        freq[index]+=1
    else:
        unique.append(num)
        freq.append(1)

for i in range(len(unique)):
    print(f"{unique[i]} - {freq[i]} times")
```

```python
def sum_of_lists(iter):
    highest_sum = 0
    largest_list:list
    for li in range(iter):
        user=input()
        li = [int(x) for x in user.split(" ")]
        sum_of_list=0
        for j in li:
            sum_of_list+=j
        if sum_of_list>highest_sum:
            highest_sum=sum_of_list
            largest_list=li
    print(highest_sum)
    print(largest_list)

num_of_lists=int(input())
sum_of_lists(num_of_lists)

```

```python
def cross_multiply(l1,l2):
    new_list = []
    for i in l1:
        for j in l2:
            new_list.append(i*j)
    return new_list

l1_entry=input()
l1=[int(x) for x in l1_entry.split(" ")]
l2_entry=input()
l2=[int(x) for x in l2_entry.split(" ")]
soln = cross_multiply(l1,l2)
print(soln)
```

```python
def is_ub(entered):
    l1=[int(x) for x in entered.split(" ")]
    is_ub_jumper=False
    sorted_num=[]
    for i in range(len(l1)-1):
        calc= l1[i+1]-l1[i]
        if calc<0:
            calc*=-1
            sorted_num.append(calc)
        else:
            sorted_num.append(calc)
    sorted_num.sort()
    if sorted_num[0]>0 and sorted_num[-1]<len(l1):
        is_ub_jumper=True


    return is_ub_jumper



output=""
while True:
    user = input()
    if user == "STOP":
        print(output)
        break
    else:
        jumper = is_ub(user)
        if jumper:
            output+="UB Jumper\n"
        else:
            output+="Not UB Jumper\n"
```

```python
user = input()
lower=[]
upper=[]
odd=[]
even=[]

for i in user:
    if i.isalpha():
        if i.islower():
            lower.append(i)
        elif i.isupper():
            upper.append(i)
    else:
        if int(i)%2==0:
            even.append(i)
        else:
            odd.append(i)
lower.sort()
upper.sort()
odd.sort()
even.sort()
sorted = "".join(lower)+"".join(upper)+"".join(odd)+"".join(even)

print(sorted)
```

```python
def team_count(limit,times):
    user = input()
    l1 = [int(x) for x in user.split()]
    if len(l1)==limit:
        teams=0
        count=0
        for i in l1:
            if i+times<=5:
                count+=1
        if count%3==0:
            teams+=(count/3)
        print(int(teams))
    else:
        print("NOT ENOUGH DATA")
limit, times = input().split()
team_count(int(limit),int(times))
```
