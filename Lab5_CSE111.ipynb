```python
class Farmer:
    def __init__(self,info=None):
        self.crops = []
        self.fish = []
        self.info = info
        if info == None:
            print("Welcome to your farm!")
        elif type(info) == int:
            print("welcome to your farm. Your farm ID is:",self.info)
        elif type(info) == str:
            print("Welcome to your farm,",self.info)
    def addCrops(self,*c):
        if c:
            self.crops.extend(c)
            print(len(self.crops),"crop(s) added")
        else:
            print("No crops added.")
    def addFishes(self,*f):
        if f:
            self.fish.extend(f)
            print(len(self.fish),"fish(s) added")
        else:
            print("No fish added")
    def showGoods(self):
        if len(self.crops) == 0:
            print("You don't have any crop(s).")
        else:
            print("You have", len(self.crops), "crop(s):", ",".join(self.crops))
        if len(self.fish) == 0:
            print("You don't have any fish(s).")

        else:
            print("You have", len(self.fish), "fish(s):", ",".join(self.fish))

f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")











```

```python
class Department:
  def __init__(self, dept="ChE Department", sections=5):
    self.dept = dept
    self.sections = sections
    print(f"The {self.dept} has {self.sections} sections.")
    self.std = []
    self.dept_lst = []
    self.total = 0
    self.std_avg = 0

  def add_students(self, *students):
    if len(students) != self.sections:
      print(f"The {self.dept} doesn't have {len(students)} sections.")
      return
    self.students = students
    for student in self.students:
      self.std.append(student)

    for std in self.std:
      self.total += std

    self.std_avg = round(self.total / self.sections, 2)
    print(f"The {self.dept} has an average of {self.std_avg} students in each section.")

  def merge_Department(self, *depts):
    self.mega_total_std = self.std_avg * self.sections
    self.mega_sect = self.sections

    for i in depts:
      print(f"{i.dept} is merged to {self.dept}.")
      self.dept_lst.append(i)

    for dept in self.dept_lst:
      self.mega_total_std += (dept.std_avg * dept.sections)

    self.mega_avg = round(self.mega_total_std / self.mega_sect, 2)

    return f"Now the {self.dept} has an average of {self.mega_avg} students in each section."


d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21, 30, 40, 36, 10, 32, 27, 51, 45, 15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
class Foodie:
    def __init__(self, name):
        self.name = name
        self.menu = menu
        self.orders = []
        self.total = 0

    def order(self, *items):
        for item in items:
            food, quant = item.split('-')
            quant = int(quant)
            if food in self.menu:
                individual_price = self.menu[food] * quant
                found = False
                for i, (existing_food, existing_quant) in enumerate(self.orders):
                    if existing_food == food:
                        self.orders[i] = (food, existing_quant + quant)
                        found = True
                        break
                if not found:
                    self.orders.append((food, quant))
                self.total += individual_price
                print(f"Ordered - {food}, quantity - {quant}, price (per Unit) - {self.menu[food]}.")
                print(f"Total price - {individual_price}")

    def pay_tips(self, tips=0):
        self.tips = tips
        if not self.tips:
            print("No tips to the waiter.")
        else:
            self.total += self.tips
            print(f"Gives {self.tips}/- tips to the waiter.")

    def show_orders(self):
        if not self.orders:
            print(f"{self.name} has 0 item(s) in the cart.")
            print("Items: []")
        else:
            items = []
            for food, quant in self.orders:
                price = self.menu[food] * quant
                items.append(f"{food}")
            print(f"{self.name} has {len(self.orders)} item(s) in the cart.")
            print(f"Items: {items}")
            print(f"Total spent: {self.total}.")
        return ""

menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}


f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())

```

```python
class Shopidify:
    def __init__(self, user=None):
        self.user= user
        print (f"Welcome {self.user} to Shopidify" if self.user else "Welcome to Shopidify")

        self.cart = []
        self.purchased = []

    def add_to_cart(self, item, quantity=1):
        if isinstance(item, list):
            for i in range(0, len(item), 2):
                self.add_to_cart(item[i], item[i + 1])
        else:
            self.cart.append((item, quantity))

    def display_cart(self):
        print(f"Items in the cart for {'Guest' if not self.user else self.user}:")
        for item, quantity in self.cart:
            print(f"- {item}: {quantity}x")

    def checkout(self):
        if self.user:
            print(f"Checkout completed for {self.user}")
        else:
            print("Checkout completed for Guest")
        self.purchased.append(list(self.cart))
        self.cart = []

    def display_history(self):
        print(f"Purchase history for {'Guest' if not self.user else self.user}:")
        for i, transaction in enumerate(self.purchased, start=1):
            print(f"Transaction {i}:")
            for item, quantity in transaction:
                 print(f"- {item}: {quantity}x")
        print()

guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
class Author:
    def __init__(self, name=""):
        self.name = name
        self.book = []

    def addBook(self, title, genre):
        if self.name == "":
            print("A book cannot be added without an author name")
        else:
            self.book.append((genre, title))

    def setName(self,n):
        self.name = n

    def printDetail(self):
        total_books = len(self.book)
        print(f"Number of Book(s): {total_books}")
        print(f"Author Name: {self.name}")

        g_books = {}
        for genre, title in self.book:
            if genre not in g_books:
                g_books[genre] = [title]
            else:
                g_books[genre].append(title)

        for genre, titles in g_books.items():
            print(f"{genre}: {', '.join(titles)}")
a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
