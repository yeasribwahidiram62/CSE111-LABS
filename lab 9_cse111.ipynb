```python
class SultansDine:
    total_branches = 0
    total_sell = 0
    branch_lst = []
    def __init__(self,branch_name):
        self.b = branch_name
        self.branch_sell = 0
        SultansDine.total_branches += 1
        SultansDine.branch_lst.append(self)


    def sellQuantity(self,quantity):
        if quantity < 10:
            self.branch_sell = quantity * 300
        elif quantity < 20:
            self.branch_sell = quantity * 350
        else:
            self.branch_sell = quantity * 400
        SultansDine.total_sell += self.branch_sell

    def branchInformation (self):
        print(f"Branch Name : {self.b}\nBranch Sell: {self.branch_sell}")
    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {cls.total_branches}\nTotal Sell: {cls.total_sell} Taka ")
        if cls.total_branches > 0 :
            t_p = sum(i.branch_sell for i in cls.branch_lst)
            for i in cls.branch_lst:
                percentage = (i.branch_sell/t_p) * 100
                print(f"Branch Name: {i.b},Branch Sell: {i.branch_sell} Taka")
                print(f"Branch consists of total sell's: {percentage:.2f}%")
SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()





```
